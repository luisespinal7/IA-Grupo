## Columnas del dataset:
Model Year: año de modelo del vehículo.

Make: marca del vehículo

Model: modelo del veículo.

Vehicle Class: clasificación del vehículo.

Engine Size(L): capacidad de los cilindros, medido en litros.

Cylinders: cantidad de cilindros en el vehículo.

Transmission: transmisión del vehículo. A = automatic; AM = automated manual; AS = automatic with select shift; AV = 
continuously.

Fuel Type: tipo de gasolina que utiliza el vehículo.

Fuel Consumption (City (L/100 km): las clasificaciones de consumo de combustible en ciudad se muestran en litros cada 100 
kilómetros (L/100 km).

Fuel Consumption(Hwy (L/100 km)): las clasificaciones de consumo de combustible en carretera se muestran en litros cada 100 
kilómetros (L/100 km)

Fuel Consumption(Comb (L/100 km)): la calificación combinada (55% ciudad, 45% carretera) se muestra en L/100 km

Fuel Consumption(Comb (mpg)): la calificación combinada (55% ciudad, 45% carretera) se muestra en millas por galón imperial 
(mpg)

CO2 Emissions(g/km): las emisiones de dióxido de carbono del tubo de escape (en gramos por kilómetro) para la conducción 
combinada en ciudad y carretera.

CO2 Rating: las emisiones del tubo de escape de dióxido de carbono calificadas en una escala de 1 (peor) a 10 (mejor)

Smog Rating: las emisiones del tubo de escape de contaminantes formadores de smog clasificados en una escala de 1 (peor) a 10 
(mejor).

```python
nan_analysis = pd.DataFrame()
```

```python
for make_ in fuel_consumption.loc[ : , "Make" ].drop_duplicates():
    makedf = fuel_consumption.loc[ fuel_consumption['Make'] == make_ , : ]    
    for column_ in makedf.columns:        
        percentage =  np.sum(pd.isna( makedf.loc[ : , column_ ]  )) / makedf.shape[0]
        dfrow = pd.DataFrame({
            'Make': [ make_ ]
            , 'column' : [ column_ ]
            , 'percentage.of.nans' : [ percentage ]
        })
        nan_analysis = pd.concat([nan_analysis, dfrow])
```
